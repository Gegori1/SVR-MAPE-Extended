```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt 
import plotly.express as px
```

# Diabetes dataset without symbolic transformation

## load data

```python
jsn_path = Path().resolve() / "Logs" / "Diabetes_mape_leo.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

## convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]]*-1, params], axis = 1)
```

## plot

```python
df1 = df.query("target < 40")
df2 = df.query("target < 35")
df3 = df.query("target < 34")
```

```python
px.scatter_3d(df2, x = "lamda", y = "gamma", z = "epsilon", color = "target", hover_data = ["C", "epsilon"])
```

```python
px.scatter(df2, x="gamma", y="epsilon", color="target")
```

```python
px.scatter_3d(df3, x = "C", y = "gamma", z = "target", color = "target", hover_data = ["C", "epsilon", "lamda", "gamma"])
```

```python
df.sort_values("target").head()
```

```python
df.sort_values("target").head(20).agg(("max", "min"))
```

# Diabetes dataset with Symbolic Transformation

## load data

```python
jsn_path = Path().resolve() / "Logs" / "Diabetes_mape_let.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

## convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]]*-1, params], axis = 1)
```

## plot

```python
df1 = df.query("target < 39")
df2 = df.query("target < 35")
df3 = df.query("target < 33.5")
```

```python
px.scatter_3d(df2, x = "lamda", y = "gamma", z = "C", color = "target", hover_data = ["epsilon"])
```

```python
px.scatter(df2, x="gamma", y="epsilon", color="target")
```

```python
px.scatter_3d(df3, x = "C", y = "epsilon", z = "target", color = "target", hover_data = ["epsilon", "C", "lamda", "gamma"])
```
