```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt 
import plotly.express as px
```

# Boston

## load data

```python
jsn_path = Path().resolve() / "Logs" / "Boston_mape_leo.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

## convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]]*-1, params], axis = 1)
```

## plot

```python
df1 = df.query("target < 13")
df2 = df.query("target < 11")
```

```python
px.scatter_3d(df2, x = "lamda", y = "gamma", z = "epsilon", color = "target", hover_data = ["C"])
```

```python
px.scatter(df1, x="gamma", y="epsilon", color="target")
```

```python
px.scatter_3d(df2, x = "C", y = "lamda", z = "target", color = "target", hover_data = ["epsilon", "C", "lamda"])
```

```python
df.sort_values("target").head(2)
```

```python
df.sort_values("target").head(20).agg(("max", "min"))
```

# Boston with Symbolic Transformation

## load data

```python
jsn_path = Path().resolve() / "Logs" / "Boston_mape_let.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

## convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]]*-1, params], axis = 1)
```

## plot

```python
df1 = df.query("target < 13")
df2 = df.query("target < 11")
df3 = df.query("target < 10.35")
```

```python
px.scatter_3d(df3, x = "lamda", y = "gamma", z = "C", color = "target", hover_data = ["epsilon"])
```

```python
px.scatter(df3, x="C", y="target", color="target", hover_data = ["epsilon"])
```

```python
px.scatter_3d(df3, x = "C", y = "epsilon", z = "target", color = "target", hover_data = ["epsilon", "C", "lamda"])
```
